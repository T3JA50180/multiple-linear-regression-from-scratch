```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from numba import jit
from tqdm.auto import tqdm
```

```python
df = pd.read_csv('Student_Performance.csv')
df.head()
```

```python
def z_score_normalize(a):
    mean = np.mean(a, axis=0)
    std = np.std(a, axis=0)
    return (a - mean) / std
```

```python
df_np = z_score_normalize(df.to_numpy())

X = df_np[:, :-1]
y = df_np[:, -1]
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
ax = ax.flatten()

for i, col in enumerate(df.columns[:-1]):
    ax[i].set_xlabel(col)
    ax[i].set_ylabel(df.columns[-1])
    ax[i].scatter(X[:, i], y, s=7)

fig.delaxes(ax[-1])
```

```python
msk = np.random.permutation(X.shape[0])
m = round(0.8 * X.shape[0])
n = X.shape[1]

X_train = X[msk[:m]]
y_train = y[msk[:m]]

X_test = X[msk[m:]]
y_test = y[msk[m:]]
```

```python
@jit
def compute_gradient(X, y, w, b):
    y_hat = np.dot(w, X.transpose()) + b

    dj_dw = np.dot((y_hat - y), X) / m
    dj_db = np.sum(y_hat - y) / m

    return dj_dw, dj_db
```

```python
@jit
def compute_cost(X, y, w, b):
    m = X.shape[0]
    y_hat = np.dot(w, X.transpose()) + b
    J = np.sum((y_hat - y) ** 2) / (2 * m)
    return J
```

```python
def gradient_descent(X, y, w, b, num_iters, alpha):
    J_hist = []
    
    for i in tqdm(range(num_iters)):
        dj_dw, dj_db = compute_gradient(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        J_hist.append(compute_cost(X, y, w, b))
        if i % (num_iters // 10) == 0:
            print(f'Iteration = {i}, Cost = {J_hist[-1]}')

    return w, b, J_hist
```

```python
w = np.zeros(n)
b = 0
num_iters = 10000000
alpha = 0.001

w, b, J = gradient_descent(X_train, y_train, w, b, num_iters, alpha)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(0, len(J)), J)
```

```python
y_test_hat = np.dot(w, X_test.transpose())
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
ax = ax.flatten()

for i, col in enumerate(df.columns[:-1]):
    ax[i].set_xlabel(col)
    ax[i].set_ylabel(df.columns[-1])
    ax[i].scatter(X_test[:, i], y_test, s=7)
    ax[i].scatter(X_test[:, i], y_test_hat, s=4)

fig.delaxes(ax[-1])
```
